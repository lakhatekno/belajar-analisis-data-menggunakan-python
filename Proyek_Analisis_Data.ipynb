### Cleaning Data

#### products_dataset
dataset products dengan dropna karena tidak mungkin mengisi missing value untuk nama-nama kategori produk.

```python
products_df.info()
```

```python
products_df.dropna(inplace=True)
products_df.info()
```

#### orders_df
dataset orders_df dengan dropna pada kolom 'approved_at' saja dan membuang kolom dengan id = [5, 6, 7] karena pada kolom lain dengan missing value tidak diperlukan pada business case yang ditanyakan.

```python
orders_df.info()
```

```python
orders_df.dropna(subset = ['order_approved_at'], inplace=True)
orders_df.drop([
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
], axis=1, inplace=True)
orders_df.info()
```

#### product category translate dan payment type method
mengganti underscore dengan spasi dan kapitalisasi tiap kata

```python
def capitalize_rm_underscore(text):
    return ' '.join(word.capitalize() for word in text.split('_'))
```

```python
product_category_translate.head()
```

```python
product_category_translate['product_category_name_english'] = product_category_translate['product_category_name_english'].apply(capitalize_rm_underscore)
product_category_translate.head()
```

```python
order_payments_df.head()
```

```python
order_payments_df['payment_type'] = order_payments_df['payment_type'].apply(capitalize_rm_underscore)
order_payments_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apa kategori produk yang paling laku terjual

```python
fig, [ax1, ax2] = plt.subplots(2, 1, figsize= [10, 10])
data_to_visualize = by_product_category.sort_values(by='order_count', ascending=False).head(5)

colors = ['#d63631', '#f2a49f', '#f2a49f', '#f2a49f', '#f2a49f']

sns.barplot(
    data = data_to_visualize,
    y = data_to_visualize.index.values,
    x = 'order_count',
    ax = ax1,
    palette=colors
)
for i, value in enumerate(data_to_visualize['order_count']):
    ax1.text(value, i, str(value), ha='left', va='center')


data_to_visualize = by_product_category.sort_values(by='total_price', ascending=False).head(5)

sns.barplot(
    data = data_to_visualize,
    y = data_to_visualize.index.values,
    x = 'total_price',
    ax = ax2,
    palette=colors
)
for i, value in enumerate(data_to_visualize['total_price']):
    ax2.text(value, i, str(value), ha='left', va='center')

ax1.set_xlabel('')
ax1.set_title('Jumlah Pesanan')

ax2.set_xlabel('')
ax2.set_title('Total Pendapatan')

fig.suptitle('Top 5 Kategori Produk')

plt.subplots_adjust(hspace=0.3) 
plt.show()
```

Jumlah pesanan kategori 'Bed Bath Table' tertinggi di antara seluruh kategori produk degnan total transaksi sebanyak 9417 kali. Jumlah itu tidak berbeda jauh dengan kategori 'Health Beauty' yang berada di posisi kedua dengan total transaksi sebanyak 8836 kali.

Sementara itu, penjualan tertinggi dicapai oleh kategori 'Health Beauty' dengan perolehan pendapatan 1,258,681.34 mata uang setempat. Pendapatan ini memiliki selisih hanya 53,675.66 dengan kategori 'Watches Gifts' di posisi kedua.

Terlihat bahwa selisih nominal dan selisih jumlah transaksi pada tingkat 5 teratas tidak terlalu jauh, sehingga bisa dikatakan bahwa penjualan antar kategori tidak terdapat ketimpangan dan memiliki pasarnya masing-masing yang hampir setara besarnya.

### Pertanyaan 2: Apa metode pembayaran yang paling populer digunakan

```python
payment_methods_to_visualize = order_payments_df.groupby(by='payment_type').agg({
    "order_id": "nunique",
    "payment_value": ["sum", "min", "max", "mean"]
})

payment_methods_to_visualize.index.names = ['Payment Method']
payment_methods_to_visualize.columns = [
    'Transaction',
    'Total Amount',
    'Minimum',
    'Maximum',
    'Average'
]

fig, [ax1, ax2] = plt.subplots(2, 1, figsize=[10, 10])

sns.barplot(
    x = payment_methods_to_visualize['Transaction'].sort_values(ascending=False),
    y = payment_methods_to_visualize['Transaction'].sort_values(ascending=False).index.values,
    ax=ax1,
    palette = colors
)

sns.barplot(
    x = payment_methods_to_visualize['Total Amount'].sort_values(ascending=False),
    y = payment_methods_to_visualize['Transaction'].sort_values(ascending=False).index.values,
    ax=ax2,
    palette = colors
)

ax1.set_title('Metode Pembayaran Terpopuler')
ax1.set_xlabel('')
for i, value in enumerate(payment_methods_to_visualize['Transaction'].sort_values(ascending=False)):
    ax1.text(value, i, str(value), ha='left', va='center')
    
ax2.set_title('Nilai Transaksi')
ax2.set_xlabel('')

for i, value in enumerate(payment_methods_to_visualize['Total Amount'].sort_values(ascending=False)):
    ax2.text(value, i, str(value), ha='left', va='center')

plt.subplots_adjust(hspace=0.3)
plt.show()
```

Metode pembayaran yang paling banyak digunakan adalah Credit Card dengan total transaksi sebanyak 76505 kali transaksi dan nominal 12,542,084.19 mata uang setempat. Selisih penggunaan metode pembayaran ini cukup signifikan dibanding metode lainnya. Debit card hanya digunakan sebanyak 1538 kali dengan nilai transaksi sebesar 217,989.79.
